```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
# enter the search item like iphone, boat, earphones, phone
search=input('Enter the search item:- ')
# flipkart url
url = 'https://www.flipkart.com/search?q'+search
```

```python
flipkart_html=BeautifulSoup(requests.get(url).text,'html')
```

```python
flipkart_html
```

```python
flipkart_html.prettify()
```

```python
bigboxes = flipkart_html.find_all('div',class_='_1AtVbE col-12-12')
del bigboxes[0:3]
box = bigboxes[0]
```

```python
productLink = "https://www.flipkart.com" + box.div.div.div.a['href']
prodRes = requests.get(productLink)
prodRes.encoding='utf-8'
prod_html = BeautifulSoup(prodRes.text, "html.parser")
# print(prod_html)
commentboxes = prod_html.find_all('div', {'class': "_16PBlm"})
```
